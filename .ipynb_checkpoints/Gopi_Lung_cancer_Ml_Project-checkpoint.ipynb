# Splitting the Dataset: Training and Testing

```python
#dependent_variable 
X = lung_cancer_data.drop('LUNG_CANCER', axis=1)
print(X)
```

```python
y = lung_cancer_data['LUNG_CANCER']
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=0)
```

# Model_1: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score,confusion_matrix
```

```python
#Fitting simple linear regression to the training test
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, y_train)
```

```python
#Predicting the test set results
LR_predicts = logistic_regression.predict(X_test)
```

```python
print("confusion matrix\n",confusion_matrix(y_true = y_test, y_pred = LR_predicts))
class_report = classification_report(y_test, LR_predicts)
print("\nclassification report\n",class_report)
```

```python
LR_confusion_matrix = confusion_matrix(y_true = y_test, y_pred = LR_predicts)
#plot_confusion_matrix(cm,level,title = "confusion_matrix")
sns.heatmap(LR_confusion_matrix, annot=True, cmap="Blues", fmt="d")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```

# Model_2: K-nearest neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors = 3)
classifier.fit(X_train, y_train)
```

```python
#Predicting the Test set result
knn_y_pred = classifier.predict(X_test)
```

```python
print("confusion matrix\n",confusion_matrix(y_test,knn_y_pred))
class_report = classification_report(y_test, knn_y_pred)
print("classification report\n",class_report)
```

```python
knn_confusion_matrix = confusion_matrix(y_test, knn_y_pred)
#plot_confusion_matrix(cm,level,title = "confusion_matrix")
sns.heatmap(knn_confusion_matrix, annot=True, cmap="Blues", fmt="d")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```
